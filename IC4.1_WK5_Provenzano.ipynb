#Task 1 Create and Insert

```python
# Create the table with STRICT types
%%sql
CREATE TABLE IF NOT EXISTS products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
) STRICT;
```

```python
#2. Insert at least 3 items (Perform a 'Create' operation)
%%sql
INSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00);
INSERT INTO products (name, price) VALUES ('Football Hat', 35.00);
INSERT INTO products (name, price) VALUES ('Team Hoodie', 70.00);
```

```python
%%sql
SELECT * FROM products;
```

#Task 2 Read and Filter

```python
#Show all products that cost more than $50
%%sql
SELECT * FROM products WHERE price > 50
```

#Task 3 Update and Delete

```python
#Update a price
%%sql

UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';
```

```python
#Delete an item
%%sql
DELETE FROM products
WHERE product_id = 1;
```
